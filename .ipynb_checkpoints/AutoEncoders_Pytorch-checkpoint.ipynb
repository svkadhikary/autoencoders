```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torchvision
%matplotlib inline
```

```python
transform = torchvision.transforms.Compose([torchvision.transforms.Resize((28, 28)),
                                            torchvision.transforms.ToTensor(),
                                           torchvision.transforms.Normalize((0.0), (1.0,),)])
trainset = torchvision.datasets.MNIST(root="./", train=True, download=True, transform=transform)
testset = torchvision.datasets.MNIST(root="./", train=False, download=True, transform=transform)
```

```python
for img, lab in trainset:
    print(img.numpy())
    break
```

```python
batch_size = 256
train_loader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(testset, batch_size=batch_size, shuffle=True)
```

```python
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True, figsize=(10, 8))
for images, labels in train_loader:
    for i in range(batch_size):
        row, col = i//4, i%4
        ax[row, col].imshow(images[i].numpy().transpose([1, 2, 0]))
        if i == 15:
            break
plt.show()
```

```python
class AutoEncoder(nn.Module):
    def __init__(self, n_hidden_1, n_hidden_2, n_encoded, n_input):
        super(AutoEncoder, self).__init__()
        self.encoding_1 = nn.Linear(n_input, n_hidden_1)
        self.encoding_2 = nn.Linear(n_hidden_1, n_hidden_2)
        self.encoding_final = nn.Linear(n_hidden_2, n_encoded)
        self.decoding_1 = nn.Linear(n_encoded, n_hidden_2)
        self.decoding_2 = nn.Linear(n_hidden_2, n_hidden_1)
        self.decoding_final = nn.Linear(n_hidden_1, n_input)
        self.sigmoid_act = nn.Sigmoid()
        self.relu_act = nn.ReLU()
        
    def encoder(self, X):
        x = self.sigmoid_act(self.encoding_1(X))
        x = self.sigmoid_act(self.encoding_2(x))
        x = self.relu_act(self.encoding_final(x))
        return x
    
    def decoder(self, enc_x):
        x = self.sigmoid_act(self.decoding_1(enc_x))
        x = self.sigmoid_act(self.decoding_2(x))
        x = self.decoding_final(x)
        return x
        
    def forward(self, inp):
        encoded = self.encoder(inp)
        decoded = self.decoder(encoded)
        return decoded
        
```

```python
def cost_fn(y_true, y_pred):
    return nn.MSELoss(reduction='mean')(y_pred, y_true)
```

```python
def train(train_loader, model, epochs, device, lr):
    opt = torch.optim.RMSprop(model.parameters(), lr=0.01, momentum=0.9)
    for epoch in range(epochs):
        epoch_loss = 0
        for i, (inp, target) in enumerate(train_loader):
            inp = inp.to(device)
            flattened_inp = inp.view(inp.size(0), -1)
            decoded = model(flattened_inp)
            opt.zero_grad()
            loss = cost_fn(decoded, flattened_inp)
            epoch_loss += loss
            loss.backward()
            opt.step()
        print(f"epoch {epoch+1} done, loss = {epoch_loss}")
            
```

```python
# initialize input shapes for data, hidden layers, batch size, epochs and learning rate
n_input = 28*28
n_hidden_1 = 512
n_hidden_2 = 256
n_encoding = 32
lr = 0.01
epochs = 20
batch_size = 256
```

```python
model = AutoEncoder(n_hidden_1, n_hidden_2, n_encoding, n_input)
device = torch.device("cuda" if torch.cuda.is_available() else 'cpu')
model.to(device)
# train the model
train(train_loader, model, epochs, device, lr)
```

```python
test_img, _ = next(iter(test_loader))
test_img = test_img.to(device)
test_img = test_img.view(test_img.size(0), -1)
encode_decode = model(test_img)
```

```python
# lets check encoded and decoded images with original images
fig, ax = plt.subplots(2, 10, sharex=True, sharey=True, figsize=(10, 2))
for i in range(10):
    ax[0][i].imshow(np.reshape(test_img[i].cpu().detach().numpy(), (28, 28)))
    ax[1][i].imshow(np.reshape(encode_decode[i].cpu().detach().numpy(), (28, 28)))
```
